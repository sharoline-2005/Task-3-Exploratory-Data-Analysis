```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Visualization settings
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# Upload netflix_titles.csv manually in Colab
df = pd.read_csv("netflix.csv")

# Display first few rows
df.head()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
top_countries = df['country'].value_counts().head(10)

sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title("Top 10 Countries by Content Count")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)

sns.lineplot(x=top_countries.values, y=top_countries.index)
plt.title("Top 10 Countries by Content Count")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.show()
```

```python
# Fill missing categorical values
for col in ['country', 'language', 'cast', 'director', 'composer writer', 'genre']:
    if col in df.columns:
        df[col] = df[col].fillna("Unknown")

# Convert numeric columns safely
df['year'] = pd.to_numeric(df['year'], errors='coerce')
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['vote'] = pd.to_numeric(df['vote'], errors='coerce')
df['runtime'] = pd.to_numeric(df['runtime'], errors='coerce')
```

```python
plt.figure()
plt.hist(df['year'].dropna(), bins=20)
plt.title("Distribution of Release Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
plt.hist(df['runtime'].dropna(), bins=20)
plt.title("Runtime Distribution")
plt.xlabel("Runtime (minutes)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
sns.countplot(data=df, x='kind')
plt.title("Content Type Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,6))
df['genre'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Genres")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,6))
df['country'].value_counts().head(10).plot(kind='barh')
plt.title("Top 10 Countries")
plt.xlabel("Count")
plt.ylabel("Country")
plt.show()
```

```python
plt.figure()
sns.boxplot(x=df['rating'].dropna())
plt.title("Rating Outliers")
plt.show()
```

```python
plt.figure()
sns.boxplot(x=df['runtime'].dropna())
plt.title("Runtime Outliers")
plt.show()
```

```python
numeric_df = df[['year', 'rating', 'vote', 'runtime']].dropna()

plt.figure()
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
